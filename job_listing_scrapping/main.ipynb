```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
endpoint = "https://realpython.github.io/fake-jobs/"
response = requests.get(endpoint)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
print(soup)
```

Get job title

```python
jobs_title = soup.find_all("h2", class_="title")
titles = []
for job in jobs_title:
    titles.append(job.text.strip())
print(titles)
```

Get Company name

```python
jobs = soup.find_all("h3", class_="company")
companies = []
for job in jobs:
    companies.append(job.text.strip())
print(companies)
```

Get location

```python
jobs = soup.find_all("p", class_="location")
locations = []
for job in jobs:
    locations.append(job.text.strip())
print(locations)
print(len(locations))
```

Get job detail urls

```python
jobs = soup.find_all("a")
urls = []
for job in jobs:
    urls.append(job.get('href'))
print(urls)
urls = list(filter(lambda x: x != "https://www.realpython.com", urls))
print(len(urls))
```

```python
header = ["title", "company", "location", "urls"]
data = [ titles, companies, locations, urls ]
transposee = list(map(list, zip(*data)))

print(transposee)
```

```python
with open("personnes.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)

    writer.writerow(header)
    writer.writerows(transposee)

print("CSV cr avec succs !")
```
